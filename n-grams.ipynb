```python
import spacy
from sklearn.feature_extraction.text import CountVectorizer
nlp=spacy.load("en_core_web_sm")
```

```python
v=CountVectorizer(ngram_range=(1,2))
v.fit(['Thor loves eating pizza'])
```

```python
v.vocabulary_
```

```python
v.get_feature_names_out()
```

```python
corpus=[
    'Iron Man loves Black Widow',
    'Thor also loves Black Widow',
    'But Black Widow loves Loki',
]
```

```python
bag=v.fit_transform(corpus)
```

```python
bag.toarray()
```

```python
v.vocabulary_
```

```python
def preprocess(text):
    filtered=[]
    doc=nlp(text)
    for token in doc:
        if token.is_stop==False and token.is_punct==False:
            filtered.append(token.lemma_)
    return " ".join(filtered)
            
    
```

```python
corpus=[preprocess(text) for text in corpus]
```

```python
corpus
```

```python
preprocess("Thor loves eating pizza")
```

```python
bag=v.fit_transform(corpus)
```
