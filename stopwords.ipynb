```python
nlp=spacy.load("en_core_web_sm")

doc=nlp("We just opened our wings, the flying part is coming soon")


for token in doc:
    if token.is_stop:
        print(token.text)
```

```python
def stop_word(text):
    doc=nlp(text)
    new=[token.text for token in doc if not token.is_stop]
    return " ".join(new)

```

```python
stop_word("We just opened our wings, the flying part is coming soon")
```
