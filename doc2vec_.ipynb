```python
import gensim
from gensim.models.doc2vec import Doc2Vec,TaggedDocument
```

```python
df=pd.read_json("Cell_Phones_and_Accessories_5.json",lines=True)
df=df.head(1000)
```

```python
document=df.reviewText
document=document.values
document=" ".join(document)
import spacy
nlp=spacy.load("en_core_web_sm")
nlp.max_length = len(document) + 1000 # add buffer for safety
```

```python
import gensim
from gensim.models.doc2vec import Doc2Vec,TaggedDocument
```

```python
doc=nlp(document)
document
```

```python
tagged_document=[TaggedDocument(words=token.text,tags=[token.i]) for token in doc]
```

```python
model=Doc2Vec(tagged_document,vector_size=100,min_count=3,workers=4,epochs=10,dm=1)
```

```python
model.build_vocab(tagged_document)
```

```python
model.corpus_count
```

```python
model.train(tagged_document,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.infer_vector(["My name"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(model.infer_vector(["You are good"]).reshape(1,-1),model.infer_vector(["You are best"]).reshape(1,-1))
```
